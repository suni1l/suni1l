```python
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
now_australia = datetime.now() # Using current time in Australia

def calculate_posted_date(posted_text):
    """Calculates the date difference from the current Australian date
    based on the posted text.
    """
    if posted_text:
        posted_text_lower = posted_text.lower()
        if "week" in posted_text_lower:
            weeks_ago = int(posted_text_lower.split()[0])
            posted_date = now_australia - timedelta(weeks=weeks_ago)
        elif "day" in posted_text_lower:
            days_ago = int(posted_text_lower.split()[0])
            posted_date = now_australia - timedelta(days=days_ago)
        elif "hour" in posted_text_lower:
            hours_ago = int(posted_text_lower.split()[0])
            posted_date = now_australia - timedelta(hours=hours_ago)
        else:
            return None  # Handle cases like "Actively Hiring" or empty strings
        return posted_date.strftime('%Y-%m-%d')
    return None

def extract_job_details(soup):
    """Extracts job details (id, name, company, posted_date, link)
    from the parsed HTML using BeautifulSoup.
    """
    job_listings = []
    job_elements = soup.find_all('li')

    for job_element in job_elements:
        job_card = job_element.find('div', class_='job-search-card')
        if job_card:
            # Extract the job ID from the data-entity-urn attribute
            data_entity_urn = job_card.get('data-entity-urn')
            job_id = data_entity_urn.split(':')[-1] if data_entity_urn else None

            title_element = job_card.find('h3', class_='base-search-card__title')
            company_element = job_card.find('h4', class_='base-search-card__subtitle')
            link_element = job_card.find('a', class_='base-card__full-link')
            date_element = job_card.find('time', class_='job-search-card__listdate')
            if not date_element:
                date_element = job_card.find('time', class_='job-search-card__listdate--new')

            job_name = title_element.text.strip() if title_element else None
            company_name = company_element.text.strip() if company_element else None
            posted_date_text = date_element.text.strip() if date_element else None
            job_link = link_element.get('href') if link_element else None

            # Calculate the posted date
            calculated_posted_date = calculate_posted_date(posted_date_text)

            job_listings.append({
                'id': job_id,
                'name': job_name,
                'company': company_name,
                'posted_date': calculated_posted_date,
                'link': job_link
            })

    return job_listings

# Example usage (assuming 'soup' is already defined):
job_data = extract_job_details(soup)
for job in job_data:
  print(job)
```

```python
import pandas as pd

df = pd.DataFrame(job_data)

def shorten_url(long_url):
    """Shortens a URL using the TinyURL API."""
    api_url = f"http://tinyurl.com/api-create.php?url={long_url}"
    try:
        response = requests.get(api_url)
        response.raise_for_status()  # Raise an exception for bad status codes
        return response.text
    except requests.exceptions.RequestException as e:
        print(f"Error shortening URL: {e}")
        return long_url  # Return the original URL in case of an error

# Assuming 'df' is your pandas DataFrame with a 'link' column
if 'link' in df.columns:
    df['short_link'] = df['link'].apply(shorten_url)
    df = df.drop(columns=['link'])  # Remove the original long link column
    df = df.rename(columns={'short_link': 'link'}) # Rename the short link column to 'link'

    #print(df)
else:
    print("The DataFrame does not contain a 'link' column.")

```

```python
import time

def extract_job_description_from_source(html_source):
    soup = BeautifulSoup(html_source, 'html.parser')
    description_div = soup.find('div', class_='description__text description__text--rich')

    if description_div:
        # Extract all text, including within nested elements
        description_text_elements = description_div.find_all(string=True, recursive=True)
        # Clean and join the text
        description = "\n".join(text.strip() for text in description_text_elements if text.strip())

        if description:  # Check if the description has any content
            return description
        else:
            return "Description is empty (even though div was found)."
    else:
        return "Description div not found."


def process_job_descriptions_from_links(datatable):
    descriptions = []
    for index, row in df.iterrows():
        job_url = row['job_description_link']
        try:
            response = requests.get(job_url)
            response.raise_for_status()  # Raise HTTPError for bad responses
            html_content = response.text
            description = extract_job_description_from_source(html_content)
            descriptions.append(description)
        except requests.exceptions.RequestException as e:
            descriptions.append(f"Error fetching {job_url}: {e}")
        except Exception as e:
            descriptions.append(f"Error processing {job_url}: {e}")
        time.sleep(10)  # Delay for 10 seconds

    df['job_description'] = descriptions
    return df

# Assuming your DataFrame is named 'df_with_links'
df_with_descriptions = process_job_descriptions_from_links(df.copy())
print(df_with_descriptions.head())
```
